```python
# загрузка аудиофайла с Google-диска

import requests
import os

current_dir = os.path.abspath('')
parent_dir = os.path.dirname(current_dir)
audio_path = os.path.join(current_dir, 'audio.mp3')


def download_file_from_cloud_storage(id, destination):
    url = "https://docs.google.com/uc?export=download"

    session = requests.Session()

    params = { 'id': id, 'confirm': 1}
    response = session.get(url, params = params, stream = True)

    save_response_content(response, destination)


def save_response_content(response, destination):
    chunk_size = 32768

    with open(destination, "wb") as f:
        for chunk in response.iter_content(chunk_size):
            if chunk:
                f.write(chunk)

download_file_from_cloud_storage('1ediCTMcNgULkMdY1AwUdRnDhUUWyq8Nx',
                                 audio_path)
```

```python
import torch
from transformers import pipeline
```

Model Card: [https://huggingface.co/openai/whisper-large-v3](https://huggingface.co/openai/whisper-large-v3)

```python
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
pipe = pipeline(
    'automatic-speech-recognition',
    model='openai/whisper-large-v3',
    chunk_length_s=30,
    device=device,
    generate_kwargs={'task':'transcribe'}
)
```

```python
prediction = pipe(audio_path,
                  batch_size=8,
                  return_timestamps=True)
```
