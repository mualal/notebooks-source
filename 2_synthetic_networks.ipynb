Посмотреть на возможности библиотек для создания, модификации и изучения структуры синтетических сетей

О библиотке networkx: https://networkx.org/

О библиотеке igraph: https://igraph.org/


```python
import networkx as nx
import itertools
import matplotlib.pyplot as plt
```

```python
network = nx.Graph()
network.add_nodes_from([1, 2, 3, 4, 5, 6])
color_list = ['gold', 'violet', 'darkorange', 'limegreen', 'red', 'blue']
for pair in itertools.combinations([1, 2, 3, 4, 5, 6], 2):
    network.add_edge(pair[0], pair[1])
fig, ax = plt.subplots()
fig.set_facecolor('white')
plt.title('Хроматическое число 6')
nx.draw_networkx(network, node_color=color_list, with_labels=False, width=1.2, style=":")
plt.box(False)
plt.tight_layout()
plt.show()
```

```python
fig.savefig('images/ChromaticNumber6.png', dpi=300, transparent=False)
```

```python
# создание неориентированного графа (networkx)
g = nx.Graph()
print(g)
```

```python
g.add_nodes_from([0, 1, 2])
g.add_edges_from([(0, 1), (1, 2)])
print(g)
nx.draw_networkx(G=g, node_size=600, node_color='brown',font_color='white', font_weight='bold', font_size=14)
```

```python
g.add_nodes_from([3, 4, 5])
g.add_edges_from([(2, 0), (2, 3), (3, 4), (4, 5), (5, 3)])
print(g)
nx.draw_networkx(G=g)
```

```python
g.remove_edges_from([(2, 3)])
nx.draw_networkx(G=g)
```

```python
g = nx.balanced_tree(r=2, h=3)
nx.draw_networkx(G=g)
```
